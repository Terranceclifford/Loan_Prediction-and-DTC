```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score
import warnings
warnings.filterwarnings('ignore')
```

```python
missing_per = (loan.isnull().sum()/614)*100
missing_per
```

```python
column_less_than5 = missing_per[missing_per<=5].index
column_less_than5
```

```python
loan.dropna(subset = column_less_than5, inplace = True)
```

```python
loan['Credit_History'].value_counts()
```

```python
loan['Credit_History'].mode()[0]
```

```python
loan['Credit_History'] = loan['Credit_History'].fillna(loan['Credit_History'].mode()[0])
```

```python
loan.isnull().sum()
```

```python
loan['Self_Employed'].value_counts()
```

```python
loan['Self_Employed'] = loan['Self_Employed'].map({'No':0,'Yes':1})
```

```python
loan['Self_Employed'].value_counts()
```

```python
loan['Loan_Status'].value_counts()
```

```python
loan['Dependents'].value_counts()
```

```python
loan['Dependents'] = loan['Dependents'].replace('3+',4)
```

```python
#dependents columns is of type object so convert it to numeric
loan['Dependents'] = pd.to_numeric(loan['Dependents'])
```

```python
loan['Gender'] = loan['Gender'].map({'Male':1,'Female':0})
```

```python
loan['Education'] = loan['Education'].map({'Graduate':1,'Not Graduate':0})
```

```python
loan['Married'] = loan['Married'].map({'Yes':1,'No':0})
```

```python
loan['Property_Area'] = loan['Property_Area'].map({'Urban':1,'Semiurban':0,'Rural':2})
```

```python
loan['Loan_Status'].value_counts()
```

```python
loan['Self_Employed'] = loan['Self_Employed'].fillna(loan['Self_Employed'].mode()[0])
```

```python
(loan.isnull().sum()/553)*100
```

```python
loan.shape
```

```python
loan.head()
```

```python
loan.drop(columns = 'Loan_ID',inplace = True)
```

```python
X = loan.drop(columns = 'Loan_Status')
Y = loan['Loan_Status']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,Y,test_size = 0.2)
```

```python
model = LogisticRegression()
```

```python
loan.isnull().sum()
```

```python
model.fit(X_train,y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
print("acurracy",accuracy_score(y_test,y_pred))
print("confusion matrix",confusion_matrix(y_test,y_pred))
print("classification rate",classification_report(y_test,y_pred))
```
